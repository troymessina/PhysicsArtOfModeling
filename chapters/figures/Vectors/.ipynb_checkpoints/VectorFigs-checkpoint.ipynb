```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# Major ticks every 20, minor ticks every 5
major_ticks = np.arange(-17, 18, 1)
minor_ticks = np.arange(-17, 18, 0.5)

ax.set_xticks(major_ticks)
#ax.set_xticks(minor_ticks, minor=True)
ax.set_yticks(major_ticks)
#ax.set_yticks(minor_ticks, minor=True)

# And a corresponding grid
#ax.grid(which='both')

# Or if you want different settings for the grids:
#ax.grid(which='minor', alpha=0.2)
ax.grid(which='major', alpha=0.5)

plt.arrow(0,0, 9, 0, width=0.1, color='r', label=r'$\vec{a}$')
plt.annotate(r'$\vec{a}$', xy=(0, 0), xytext=(5, 0.5))
plt.arrow(0,0, 7.07, 7.07, width=0.1, color='b', label=r'$\vec{b}$')
plt.annotate(r'$\vec{b}$', xy=(0, 0), xytext=(3, 5))
plt.arrow(5,11, 10, 0, width=0.1, color='b', label=r'$\vec{c}$')
plt.annotate(r'$\vec{c}$', xy=(0, 0), xytext=(10, 12))
plt.arrow(0,13, -7.07, -7.07, width=0.1, color='b', label=r'$\vec{d}$')
plt.annotate(r'$\vec{d}$', xy=(0, 0), xytext=(-4.5, 10))
plt.arrow(-9,4, -5, 6, width=0.1, color='b', label=r'$\vec{e}$')
plt.annotate(r'$\vec{e}$', xy=(0, 0), xytext=(-11, 8))
plt.arrow(-2,0, -10, 0, width=0.1, color='b', label=r'$\vec{f}$')
plt.annotate(r'$\vec{f}$', xy=(0, 0), xytext=(-7, 0.5))
plt.arrow(-8,-12, 0, 10, width=0.1, color='b', label=r'$\vec{g}$')
plt.annotate(r'$\vec{g}$', xy=(0, 0), xytext=(-7, -8))
plt.arrow(4.5,-7, 5.5, 0, width=0.1, color='b', label=r'$\vec{h}$')
plt.annotate(r'$\vec{h}$', xy=(0, 0), xytext=(7, -6))
plt.arrow(7,-15, 7.07, 7.07, width=0.1, color='b', label=r'$\vec{i}$')
plt.annotate(r'$\vec{i}$', xy=(0, 0), xytext=(9.5, -11))
plt.grid(True)
plt.xlim(-17, 17)
plt.ylim(-17, 17)
plt.xticks(color='w')
plt.yticks(color='w')
plt.savefig('Prob3.2.png')
plt.show()
```
