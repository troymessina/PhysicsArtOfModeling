```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 10, 101)
x=np.zeros(101)
x[0:21] = 3*t[0:21]
x[21:50] = 6.0
x[50:61] = -4*(t[50:61]-5)+6
x[61:100] = -0.5*(t[61:100]-6)+2

plt.plot(t, x)
plt.xlabel('Time, t (s)')
plt.ylabel('Position, x (m)')
plt.grid(True)
plt.savefig('motionDiagram.png')
plt.show()
```

```python
plt.plot(t, x)
plt.xlabel('Time, t (s)')
plt.ylabel('Position, x (m)')
plt.grid(True)
plt.annotate('a', xy=(0, 3), xytext=(0.4, 3.1), xycoords='data')
plt.annotate('b', xy=(3, 5), xytext=(3.5, 5.7), xycoords='data')
plt.annotate('c', xy=(6, 4.5), xytext=(5.6, 4.5), xycoords='data')
plt.annotate('d', xy=(8, 1.5), xytext=(7.5, 1.4), xycoords='data')
plt.savefig('motionDiagramSoln.png')
plt.show()
```

```python
t = np.linspace(0, 10, 11)
a = -0.05
vo = 0.5
xo = 0.5
vt = vo + a*t
xt = xo + vo*t + 1/2*a*t**2

plt.plot(t, vt, 'or')
plt.xlabel(r'$t$ (s)')
plt.ylabel(r'$v_x(t)$ (m/s)')
plt.grid(True)
plt.savefig('velVst_Ex4.3.png')
plt.show()

plt.plot(t, xt, 'ob')
plt.xlabel(r'$t$ (s)')
plt.ylabel(r'$x(t)$ (m/s)')
plt.grid(True)
plt.savefig('posVst_Ex4.3.png')
plt.show()
```
