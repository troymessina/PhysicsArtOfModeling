```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, np.pi*2, 500)
halft = np.sin(0.5*x) + 6
halfb = -np.sin(0.5*x) + 6
fullt = np.sin(1*x) + 3
fullb = -np.sin(1*x) + 3
threehalft = np.sin(1.5*x) + 0
threehalfb = -np.sin(1.5*x) + 0

plt.plot(x, halft, '-k')
plt.plot(x, halfb, '--k')
plt.plot(x, fullt, '-k')
plt.plot(x, fullb, '--k')
plt.plot(x, threehalft, '-k')
plt.plot(x, threehalfb, '--k')
#plt.arrow(0,6, 6, 0)
plt.annotate(r'$L=1/2\lambda$', xy=(0,6), xytext=(2.8,6.25), xycoords='data')
plt.annotate('', xy=(0.14,6), xytext=(6,6), xycoords='data', arrowprops=dict(arrowstyle='<->', color='blue'))
plt.annotate(r'$L=\lambda$', xy=(0,6), xytext=(2.85,3.4), xycoords='data')
plt.annotate('', xy=(0.14,3), xytext=(6,3), xycoords='data', arrowprops=dict(arrowstyle='<->', color='blue'))
plt.annotate(r'$L=3/2\lambda$', xy=(0,6), xytext=(2.75,0.25), xycoords='data')
plt.annotate('', xy=(0.14,0), xytext=(6,0), xycoords='data', arrowprops=dict(arrowstyle='<->', color='blue'))
plt.axis('off')
plt.show()
```

```python
z = np.linspace(0, 3, 1000)
v = 345
f = 345
wl = v/f
wv1 = 0.2*np.sin(2*np.pi/wl*z)
wv2 = 0.2*np.sin(2*np.pi/wl*z-2*np.pi*f*0.0015)

plt.plot(z, wv1, '-b', label='t=0.0000 s')
plt.plot(z, wv2, '-.g', label='t=0.0015 s')
plt.grid(True)
plt.legend(loc=1)
plt.hlines(0, 0, 3.75, color='black')
plt.vlines(0, -2.5, 2.5, color='black')
plt.xlim(0, 3.75)
plt.ylim(-0.225, 0.24)
plt.xlabel('x (m)')
plt.ylabel('Relative Pressure')
plt.annotate('wave speed', xy=(0.25,0.225), xytext=(0.75,0.22), xycoords='data', arrowprops=dict(arrowstyle='<-', color='black'))
plt.show()
```
